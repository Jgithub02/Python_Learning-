### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# Write your code here

x1 = int(input("enter x1: "))
y1 = int(input("enter y1: "))
x2 = int(input("enter x2: "))
y2 = int(input("enter y2: "))

distance = ((x2-x1)**2 + (y2-y1)**2)**0.5

print("Distance between two coordinates: ", distance)
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Write your code here
heads = int(input("enter number of heads: "))
legs = int(input("enter number of legs: "))

# legs =  4*dogs + 2*chicken ----(1)
# heads = dogs + chicken -----(2)
# Now Eqn. (1) - (2)
# legs - heads = 3dogs + chicken-----(3)
# From Eqn. (2)
# chicken = heads - dogs-----(4)
# From eqn. (3) & (4)
# legs - heads = 3dogs + heads - dogs
# legs - 2heads = 2dogs
# dogs = (legs - 2heads)/2

dogs = (legs - 2*heads)/2
chicken = heads - dogs  # from eqn. (4)

print("Dogs: ", dogs)
print("Chicken: ", chicken)


````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# Write your code here
n = float(input("enter the number n: "))

#Sum of square of first n natural numbers = n(n+1)(2n+1)/6
sum = (n*(n+1)*(2*n+1))//6

print("Sum of Square of First n Natural Numbers: ", sum)

````
